# 使用PCA对数据进行降维

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris 
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
from sklearn import decomposition
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns = iris.feature_names)
df_iris.head()
```

```python
# n_components=1
pca = PCA(n_components=1)
pca.fit(df_iris)
print('explained_variance_ration: ',pca.explained_varience_ratio)
print('explained_varience: ',pca.explained_varience_)
print('n_components',pca.n_components_)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition

X = iris.data
y = iris.target

# PCA降维度
pca = decomposition.PCA(n_components=3)  #降到三维 
pca.fit(X)
X = pca.transform(X)
#3D图
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()

for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# 重新排序标签，使其颜色与群集结果匹配  
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.set_xlabel('第一特征向量')
ax.set_ylabel('第二特征向量')
ax.set_zlabel('第三特征向量')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
